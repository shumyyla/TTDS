```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [20, 15, 25, 10]

# Create bar chart
plt.scatter(labels, values, marker='s')

# Add labels and title
plt.xlabel('Fruit')
plt.ylabel('Quantity')
plt.title('Fruit Sales')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.normal(0, 1, 1000)

# Create histogram
plt.hist(data, bins=20, color='blue')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Show plot
plt.show()
```

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]


fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot()

ax.plot(years, chemical_exports, 'o-g', label="Chemicals",ls=':')
ax.plot(years, medicine_exports, 'x-r', label="Medicines",ls='-')

ax.set_xlabel("Years")
ax.set_ylabel("Amount (Million US$)")
xvals = np.arange(2011, 2022, 2)
yvals = np.linspace(0.80, 0.98, 10)
ax.set_xticks(xvals)
ax.set_yticks(yvals)
plt.title("LCI Exports in last 12 years")
plt.grid(True)
plt.annotate('peak of chem export',xy=(2016,0.95))
plt.legend(loc='best')
plt.style.use('default')

plt.show()
plt.savefig("mychart.png")
```

```python
from matplotlib import pyplot as plt
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]

fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1)
ax1.plot(years, chemical_exports, label='Chemicals', marker='o', c='b', ls='-', lw=2) 
ax1.set_title("LCI Chemical exports in last 12 years")
ax1.set_xlabel("Years")
ax1.set_ylabel("Amount (Million Rs.)")

ax2.plot(years, medicine_exports, label='Medicines', marker='x', c='r', ls=':', lw=2) 
ax2.set_title("LCI Medicine exports in last 12 years")
ax2.set_xlabel("Years")
ax2.set_ylabel("Amount (Million Rs.)")

plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]


fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(years, chemical_exports, label='Chemicals', marker='o', c='b', ls='-', lw=2) 
ax1.set_title("LCI Chemical exports in last 12 years")
ax1.set_xlabel("Years")
ax1.set_ylabel("Amount (Million Rs.)")

ax2.plot(years, medicine_exports, label='Medicines', marker='x', c='r', ls=':', lw=2) 
ax2.set_title("LCI Medicine exports in last 12 years")
ax2.set_xlabel("Years")
ax2.set_ylabel("Amount (Million Rs.)")

plt.tight_layout()
plt.show()
```
