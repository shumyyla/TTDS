```python
import numpy as np
# creating a dataset containing list of students and their marks
students = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
english = [85, 45, 60, 70, 83, 35]
maths=[80,95,67,55,90,40]

# Create figure and axes objects
fig = plt.figure()
ax = fig.add_subplot()
index =  np.arange(len(students))
#Draw bar plot between names (categorical variable) and marks (numerical variable) of students
ax.bar(x=index-0.22, height=english, width=0.45)
ax.bar(x=index+0.22, height=maths,width=0.45)
# adding labels
ax.set_xlabel("Names of Students")
ax.set_ylabel("Marks of Students")

# adding title
plt.title("Distribution of Student Marks")
ax.set_xticks(ticks=index, labels=students, color='g')

#plt.savefig("mybargraph.png")
# show plot
plt.show()
```

```python
# creating a dataset containing list of students and their marks
students = ["Hadeed", "Maaz", "Mujahid", "Mohid", "Kamal",  "Jamal"]
english = [85, 45, 60, 70, 83, 35]
maths = [ 80, 59, 80, 60, 65, 60]

# Create figure and axes objects
fig = plt.figure()
ax = fig.add_subplot()

index = np.arange(len(students))

#Draw bar plot between names (categorical variable) and marks (numerical variable) of students
ax.barh(y=index-0.15, width=english, height=0.3, label='English')
ax.barh(y=index+0.15, width=maths, height=0.3, label='Maths')

# adding labels
ax.set_xlabel("Marks of Students")
ax.set_ylabel("Names of Students")


plt.yticks(ticks=index, labels=students, color='g')

# adding title
plt.title("Distribution of Student Marks")

plt.legend()

plt.savefig("hbar.png")

# show plot
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

# Creating list of random values
l1 = np.random.randint(0, 100, 20)
l2 = np.random.randint(0, 100, 20)
sizes = l2.astype(int)
colors = sizes

# Create figure and axes objects
fig, ax = plt.subplots()

# Draw Scatter plot where two points intersect in l1 and l2
plt.scatter(l1,l2, s=sizes, c = colors, alpha=0.5, edgecolor='black', lw=2)

# adding labels
ax.set_xlabel("List 1")
ax.set_ylabel("List 2")
cbar = plt.colorbar()
cbar.set_label("Price of house")
# adding title
plt.title("Scatter Plot")

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np


mks = np.array([65, 85, 72,45, 87, 68, 61, 80, 77, 73, 66, 200])
# Create figure and axes objects
fig, ax = plt.subplots()

ax.boxplot(x=mks, patch_artist=True)

# title and axis labels
ax.set_title('Box Plot for Subject Marks')
ax.set_xlabel('Data Science')
ax.set_ylabel('Marks')


# add horizontal grid lines
ax.yaxis.grid(True)

print("Mean: ", np.median(mks))
print("Min: ", np.max(mks))
print("Min: ", np.min(mks))
q1, q2, q3 = np.quantile(mks,[.25, .5, .75])
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
plt.show()
```

```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import numpy as np


mks = np.array([65, 85, 72,45, 87, 68, 61, 80, 77, 73, 66, 200])
# Create figure and axes objects
fig, ax = plt.subplots()
quantiles= [0.25,0.75]
ax.violinplot(dataset=mks, showmedians=True, quantiles=quantiles)

# title and axis labels
ax.set_title('Box Plot for Subject Marks')
ax.set_xlabel('Data Science')
ax.set_ylabel('Marks')


# add horizontal grid lines
ax.yaxis.grid(True)

print("Mean: ", np.median(mks))
print("Min: ", np.max(mks))
print("Min: ", np.min(mks))
q1, q2, q3 = np.quantile(mks,[.25, .5, .75])
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
plt.show()
```
